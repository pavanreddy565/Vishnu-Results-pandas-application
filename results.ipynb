# Assigning Branches


```python
branchs={'01':'CIVIL','02':'MECH','03':'EEE','04':'ECE',"05":"CSE","45":"AIDS","42":"AIML","57":"CSBS"}
```

```python
df1["Branch"] = df1['HT.No.'].str[6:8].map(branchs)
df1['Credits'] = df1['Credits'].replace('--', 0)
df1.head()
```

```python
subjects=selected_df['Subject Name'].unique()
selected_df['Subject Name'].nunique()
```

# Subjects credits calculation


```python

subject_credits_dict = {}


for index, row in selected_df.iterrows():
    subject_name = row['Subject Name']
    credits = row['Credits']


    if subject_name not in subject_credits_dict and not pd.isnull(credits):
        subject_credits_dict[subject_name] = credits
    elif subject_credits_dict[subject_name]!=credits and credits!=0:
        subject_credits_dict[subject_name] = credits
sub=subject_credits_dict.copy()
subject_credits_dict.update({'dummy': 0})


print(sub)
```

# pivoting the table

```python
result_df = selected_df.pivot_table(index='HT.No.', columns='Subject Name', values='Grade', aggfunc=lambda x: ', '.join(x.astype(str)))
result_df.reset_index(inplace=True)
result_df.head(2)
```

# CGPA calculator funtion


```python
def sgpaCalculator(a):
    sum = 0
    credits = 0
    l = {'o': 10, 's': 9, 'a': 8, 'b': 7, 'c': 6, 'd': 5, 'f': 0, 'p': 1,'-ab-':0}

    for subject, grade in a.items():
        grade = str(grade).lower()
        if grade == 'nan' or grade == 'f' or grade =='-ab-':
            grade = 'f'
        sum += subject_credits_dict[subject] * l[grade]
        if grade != 'f':
            credits += subject_credits_dict[subject]

    if credits == 0:
        return "0.00"
    else:
       return "{:.2f}".format(sum / credits)
```

```python
result_df.insert(loc=1, column='SGPA', value=0.0)

for index, row in result_df.iterrows():
    a = {}
    for subject in sub.keys():
        a[subject] = row[subject]
    value = sgpaCalculator(a)
    result_df.at[index, 'SGPA'] = value
```

```python

def color_grade(grade):
    if isinstance(grade, str) and 'F' in grade:
        return 'color: red'
    else:
        return ''
```

# Enter your Section starting and ending rollnumbers to get results for you section

```python

start_value=input("Enter Starting Roll NO\n").upper()
end_value=input("Enter Ending Roll NO\n").upper()

le_start=input("Enter LE Starting Roll NO\n").upper()
le_end=input("Enter LE Ending Roll NO\n").upper()

selected_rows = result_df[((result_df['HT.No.'] >= start_value)& (result_df['HT.No.'] <= end_value))|(result_df['HT.No.'] >= le_start)& (result_df['HT.No.'] <= le_end)]
selected_rows=selected_rows.style.applymap(color_grade, subset=pd.IndexSlice[:, result_df.columns[2:]])

selected_rows
```
